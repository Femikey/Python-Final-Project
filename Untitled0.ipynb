```python
"""
Google Trends через него делаю анализ запросов(топи спрос ищу) на обучение зарубежом
"""

from pytrends.request import TrendReq
import pandas as pd
import matplotlib.pyplot as plt
import time

pytrends = TrendReq(hl = 'ru', tz = 180)  # tz = 180 для Москвы

# Сначала запросы на визы за последние 12 месяцев
visa_keywords = [
    'оформление визы',
    'помощь с визой',
    'виза в Европу',
    'студенческая виза',
    'виза самостоятельно'
]

pytrends.build_payload(visa_keywords, cat = 0, timeframe = 'today 12-m', geo = 'RU')
visa_interest = pytrends.interest_over_time()

if not visa_interest.empty:
    visa_interest.drop(columns = ['isPartial'], inplace = True)

    # график
    fig, ax = plt.subplots(figsize = (15, 5))
    visa_interest.plot(ax = ax)
    ax.set_title('Динамика поисковых запросов по визам', fontsize = 15)
    ax.set_xlabel('Дата')
    ax.set_ylabel('Относительный интерес')
    ax.legend(bbox_to_anchor = (1, 1), loc = 'upper left')
    plt.tight_layout()
    plt.savefig('visa_trends.png', dpi = 150, bbox_inches = 'tight')
    plt.show()

    # сохраним данные
    visa_interest.to_csv('visa_search_trends.csv', encoding = 'utf-8-sig')
    print("Средние значения интереса по запросам:")
    print(visa_interest.mean().sort_values(ascending = False))


# кол во запросов по обучению заграницей
education_keywords = [
    'обучение за рубежом',
    'поступление в зарубежный вуз',
    'учиться в Европе',
    'магистратура за границей',
    'стипендия за рубежом'
]

pytrends.build_payload(education_keywords, cat = 0, timeframe = 'today 12-m', geo = 'RU')
edu_interest = pytrends.interest_over_time()

if not edu_interest.empty:
    edu_interest.drop(columns = ['isPartial'], inplace = True)

    fig, ax = plt.subplots(figsize = (15, 5))
    edu_interest.plot(ax = ax)
    ax.set_title('Динамика запросов по обучению за рубежом', fontsize = 14)
    ax.set_xlabel('Дата')
    ax.set_ylabel('Относительный интерес')
    ax.legend(bbox_to_anchor  = (1, 1), loc = 'upper left')
    plt.tight_layout()
    plt.savefig('education_trends.png', dpi = 150, bbox_inches = 'tight')
    plt.show()

    edu_interest.to_csv('education_search_trends.csv', encoding = 'utf-8-sig')



# из каких регионов ищут больше всего
pytrends.build_payload(['обучение за рубежом'], cat = 0, timeframe = 'today 12-m', geo = 'RU')
regional = pytrends.interest_by_region(resolution = 'REGION', inc_low_vol = True, inc_geo_code = False)
regional = regional.sort_values('обучение за рубежом', ascending = False)

print("топ регионов")
print(regional.head(10))
regional.to_csv('regional_interest.csv', encoding = 'utf-8-sig')

# похожие и связанные запросы
pytrends.build_payload(['студенческая виза'], cat = 0, timeframe = 'today 12-m', geo = 'RU')
related = pytrends.related_queries()

if related['студенческая виза']['rising'] is not None:
    rising_queries = related['студенческая виза']['rising']
    print(" связанные запросы")
    print(rising_queries.to_string())
    rising_queries.to_csv('rising_queries_visa.csv', encoding = 'utf-8-sig', index = False)

if related['студенческая виза']['top'] is not None:
    top_queries = related['студенческая виза']['top']
    print("типо топ связанных запросов")
    print(top_queries.to_string())
    top_queries.to_csv('top_queries_visa.csv', encoding = 'utf-8-sig', index = False)

# сравнение стран снг
cis_countries = {
    'KZ': 'Казахстан',
    'UZ': 'Узбекистан',
    'BY': 'Беларусь',
    'KG': 'Кыргызстан',
    'RU': 'Россия'
}

cis_data = {}
for geo_code, country_name in cis_countries.items():
    try:
        pytrends.build_payload(['study abroad'], cat = 0, timeframe = 'today 12-m', geo = geo_code)
        data = pytrends.interest_over_time()
        if not data.empty:
            cis_data[country_name] = data['study abroad'].mean()
        time.sleep(5)
    except Exception as e:
        print(f"Ошибка для {country_name}: {e}")

if cis_data:
    cis_df = pd.DataFrame.from_dict(cis_data, orient = 'index', columns = ['средний интерес'])
    cis_df = cis_df.sort_values('средний интерес', ascending = False)
    print("интерес именно по странам снг")
    print(cis_df)

    fig, ax = plt.subplots(figsize = (10, 5))
    cis_df.plot(kind = 'bar', ax = ax, legend = False, color = 'steelblue')
    ax.set_title('Интерес по странам СНГ')
    ax.set_ylabel('средний индекс интереса')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.savefig('cis_interest.png', dpi=150)
    plt.show()
```

```python
# Данные по Росстату (господи спасибо что хоть ты не делаешь защиту от парсинга, просто лучший сайт, а то я так умирала пока пыталась сделать хоть чтото с авито и профи ру, но у меня не вышло((((())))))



import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import re

matplotlib.rcParams['font.family'] = 'DejaVu Sans'

def parse_spros_sheet(filepath, sheet_name, tourism_type):
    df = pd.read_excel(filepath, sheet_name = sheet_name, header = None, engine = 'openpyxl')

    rows = []
    current_year = None

    quarter_map = {
        'i квартал':   1,
        'ii квартал':  2,
        'iii квартал': 3,
        'iv квартал':  4,
    }

    for i in range(len(df)):
        cell = str(df.iloc[i, 0]).strip()
        cell_lower = cell.lower()

        year_match = re.match(r'^(20\d{2})\s*г\.?$', cell)
        if year_match:
            current_year = int(year_match.group(1))
            continue

        if cell_lower in quarter_map and current_year is not None:
            q = quarter_map[cell_lower]

            try:
                inc  = float(df.iloc[i, 1])
                same = float(df.iloc[i, 2])
                dec  = float(df.iloc[i, 3])
                bal  = float(df.iloc[i, 4])
            except (ValueError, TypeError):
                continue

            rows.append({
                'Тип_туризма':   tourism_type,
                'Год':           current_year,
                'Квартал':       q,
                'Период':        f'{current_year}Q{q}',
                'Увеличение':    inc,
                'Без_изменений': same,
                'Уменьшение':    dec,
                'Баланс':        bal,
            })

    return pd.DataFrame(rows)

sheets_map = {
    '1': 'Въездной',
    '2': 'Выездной',
    '3': 'Внутренний',
}

all_data = []
for sheet_name, tour_type in sheets_map.items():
    df_part = parse_spros_sheet('rosstat_spros.xlsx', sheet_name, tour_type)
    all_data.append(df_part)
    print(f"✓ {tour_type}: {len(df_part)} строк")

df_spros = pd.concat(all_data, ignore_index = True)

# беру данные только с 2020 года
df_spros = df_spros[df_spros['Год'] >= 2020].reset_index(drop = True)

df_spros.to_csv('rosstat_spros_clean.csv', index = False, encoding = 'utf-8-sig')



# делаю анализ и сравнение

# Средний баланс по типам туризма
avg_by_type = df_spros.groupby('Тип_туризма')['Баланс'].agg(['mean', 'min', 'max']).round(1)
print("\nСредний баланс оценок по типам:")
print(avg_by_type.to_string())

# Средний баланс по годам и типам
pivot_year = df_spros.pivot_table(
    index='Год',
    columns='Тип_туризма',
    values='Баланс',
    aggfunc='mean'
).round(1)

print("\nсредний баланс по годам")
print(pivot_year.to_string())

# Худший и лучший период для выездного
df_out = df_spros[df_spros['Тип_туризма'] == 'Выездной']
worst = df_out.loc[df_out['Баланс'].idxmin()]
best  = df_out.loc[df_out['Баланс'].idxmax()]

print(f"\nвыездной туризм")
print(f"худший период  {worst['Период']} (баланс {worst['Баланс']:+.0f})")
print(f"лучший период {best['Период']} (баланс {best['Баланс']:+.0f})")

neg_share = (df_out['Баланс'] < 0).sum() / len(df_out) * 100
print(f"доля кварталов со снижением спроса {neg_share:.0f}%")



# граыики

fig, axes = plt.subplots(2, 2, figsize=(18, 12))
fig.suptitle('Спрос на туристические услуги (Росстат, с 2020 г.)', fontsize = 15)

colors_map = {
    'Въездной':   '#3498db',
    'Выездной':   '#e74c3c',
    'Внутренний': '#27ae60',
}

ax = axes[0, 0]

for tour_type, color in colors_map.items():
    sub = df_spros[df_spros['Тип_туризма'] == tour_type]
    ax.plot(sub['Период'], sub['Баланс'], 'o-', label = tour_type, color = color, markersize = 5)

ax.axhline(0, color = 'black', linewidth = 0.8)
ax.set_title('Баланс спроса по кварталам (все типы туризма)', fontsize = 15)
ax.set_ylabel('Баланс оценок')
ax.tick_params(axis = 'x', rotation = 90, labelsize = 7)
ax.legend(fontsize = 10)
ax.grid(True, alpha = 0.3)



ax = axes[0, 1]

pivot_year.plot(kind = 'bar', ax = ax, color=[colors_map[c] for c in pivot_year.columns],
                alpha=0.85, edgecolor='white', width=0.75)

ax.axhline(0, color = 'black', linewidth = 0.8)
ax.set_title('Средний баланс спроса по годам', fontsize=12)
ax.set_ylabel('Средний баланс')
ax.set_xlabel('Год')
ax.tick_params(axis = 'x', rotation = 0)
ax.legend(fontsize = 10, title = 'Тип туризма')
ax.grid(True, alpha = 0.3, axis = 'y')

# Подписи значений
for container in ax.containers:
    for bar in container:
        val = bar.get_height()
        if val != 0:
            va = 'bottom' if val >= 0 else 'top'
            offset = 0.5 if val >= 0 else -0.5
            ax.text(bar.get_x() + bar.get_width()/2, val + offset, f'{val:+.0f}', ha = 'center', va = va, fontsize = 5)



ax = axes[1, 0]

df_out = df_spros[df_spros['Тип_туризма'] == 'Выездной']
bar_colors = ['#27ae60' if v >= 0 else '#e74c3c' for v in df_out['Баланс']]

bars = ax.bar(df_out['Период'], df_out['Баланс'], color = bar_colors, alpha = 0.85, edgecolor = 'white')
ax.axhline(0, color = 'black', linewidth = 0.8)

for bar, val in zip(bars, df_out['Баланс']):
    va = 'bottom' if val >= 0 else 'top'
    offset = 1.5 if val >= 0 else -1.5
    ax.text(bar.get_x() + bar.get_width()/2, val + offset, f'{val:+.0f}', ha = 'center', va = va, fontsize = 7)

ax.set_title('Спрос на выездной туризм по кварталам', fontsize = 12)
ax.set_ylabel('Баланс оценок')
ax.tick_params(axis = 'x', rotation = 90, labelsize = 7)
ax.grid(True, alpha = 0.3, axis = 'y')




ax = axes[1, 1]

df_out_avg = df_out.groupby('Год')[['Увеличение', 'Без_изменений', 'Уменьшение']].mean()

df_out_avg.plot(kind = 'bar', stacked = True, ax = ax, color=['#27ae60', '#95a5a6', '#e74c3c'], alpha = 0.85, edgecolor = 'white', width = 0.7)

ax.set_title('Структура ответов по выездному туризму(средние % по году)', fontsize = 12)
ax.set_ylabel('% организаций')
ax.set_xlabel('Год')
ax.tick_params(axis = 'x')
ax.legend(fontsize = 9, loc = 'lower right')
ax.grid(True, alpha = 0.3, axis = 'y')

# Подписи внутри столбцов
for container in ax.containers:
    for bar in container:
        h = bar.get_height()
        if h > 5:
            ax.text(bar.get_x() + bar.get_width()/2, bar.get_y() + h/2, f'{h:.0f}%',
                   ha = 'center', va = 'center', fontsize = 8, color = 'white')

plt.tight_layout()
plt.savefig('analysis_spros.png', dpi=150, bbox_inches='tight')
plt.show()


# какие получили выводы

avg_out = df_spros[df_spros['Тип_туризма'] == 'Выездной']['Баланс'].mean()
avg_in  = df_spros[df_spros['Тип_туризма'] == 'Внутренний']['Баланс'].mean()

# Динамика последних лет
last_2_years = df_spros[df_spros['Год'] >= df_spros['Год'].max() - 1]
trend_out = last_2_years[last_2_years['Тип_туризма'] == 'Выездной']['Баланс'].mean()

print(f""" Сравнение типов туризма (средний баланс с 2020 г.):
   Выездной:    {avg_out:+.1f}
   Внутренний:  {avg_in:+.1f}
   Разрыв: {avg_in - avg_out:+.1f} в пользу внутреннего

Состояние рынка:
   {neg_share:.0f}% кварталов с 2020 года показали снижение спроса
   Худший момент: {worst['Период']}
   Лучший момент: {best['Период']}
   Тренд последних 2 лет: {trend_out:+.1f}

Что это значит:
   Спрос на выездной туризм нестабилен (ну очев как бы)
   НО! Поток уезжающих учиться/работать (а не туризм) остаётся
     устойчивым(а это и есть то что нам нужно)
   Конкуренция за туристов слабеет (значит нам лучше и больше достанется)
""")
```
